```python
# Import sentence transformer library and choose model to make embeddings of ahadith
from sentence_transformers import SentenceTransformer,util
model = SentenceTransformer('all-MiniLM-L6-v2')
#sentences=[documents]
#embeddings = model.encode(sentences)
#embeddings.shape
```

```python
# Import csv library
import csv
```

```python
# Load saved csv file to make a format of data to be stored in Chromadb collection
with open('first_100_rows.csv') as file:
    lines = csv.reader(file)

    documents_b_m = []
    embeddings_b_m = []
    metadatas_b_m = []
    ids_b_m = []
    id = 1

    for i, line in enumerate(lines):
        if i==0:
            continue

        #Code to store ahadith in variables with embeddings, metadatas and ids
        documents_b_m.append(line[2])
        embeddings_b_m.append(model.encode(line[2]).tolist())
        metadatas_b_m.append({"hadith_id": line[1],"source": line[0]})
        ids_b_m.append(str(id))
        id+=1
```

```python
# Check stored embeddings in variable
embeddings_b_m
```

```python
documents_b_m
```

```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection_name = 'first_hundred_hadith'
```

```python
collection = client.create_collection(collection_name)
```

```python
# Storing data from variables to chromadb collection
collection.add(
    documents=documents_b_m,
    metadatas=metadatas_b_m,
    ids=ids_b_m,
    embeddings=embeddings_b_m
)
```

```python
# Giving hadith in query to get 10 similar ahadith from chromadb collection
results = collection.query(
    query_texts=['''Narrated Anas bin Malik:                     The Prophet came to Khaibar and when Allah made him victorious and he conquered the town by breaking the enemy's defense, the beauty of Safiya bint Huyai bin Akhtab was mentioned to him and her husband had been killed while she was a bride. Allah's Apostle selected her for himself and he set out in her company till he reached Sadd-ar-Rawha' where her menses were over and he married her. Then Hais (a kind of meal) was prepared and served on a small leather sheet (used for serving meals). Allah's Apostle then said to me, "Inform those who are around you (about the wedding banquet)." So that was the marriage banquet given by Allah's Apostle for (his marriage with) Safiya. After that we proceeded to Medina and I saw that Allah's Apostle was covering her with a cloak while she was behind him. Then he would sit beside his camel and let Safiya put her feet on his knees to ride (the camel).'''],
    n_results=10
)


print(results)
```

```python
# Making a query of a hadith and making its embedding
query='''Ibn Shibab reported:                      'Umar b. 'Abd al-'Aziz deferred the afternoon prayer somewhat and 'Urwa said to him: Gabriel came down and he led the Messenger of Allah (may peace be upon him) in prayer. 'Umar said to him: O 'Urwa, are you aware of what you are saying? Upon this he ('Urwa) said: I heard Bashir b. Abu Mas'ud say that he heard Abu Mas'ud say that he heard the Messenger of Allah (may peace be upon him) say: Gabriel came down and acted as my Imam, then I prayed with him, then I prayed with him, then I prayed with him. then I prayed with him. then I prayed with him. reckoning with his fingers five times of prayer.'''
query_embedding = model.encode(query)

query_embedding
```

```python
# Making a cosine similarity matrix of given hadith and stored ahadith locally(not in chromadb)
hadith_cos_sim = util.cos_sim(query_embedding,embeddings_b_m)

hadith_cos_sim
```

```python
# Converting cosine similarity matrix to an array
all_ahadith_sim = []
for i in range(hadith_cos_sim.shape[1]-1):
    all_ahadith_sim.append((hadith_cos_sim[0][i],i))

all_ahadith_sim
```

```python
# Sort array in descending order with respect to similarity quotient
all_ahadith_sim_sorted = sorted(all_ahadith_sim, key=lambda x: x[0],reverse=True)
# all_ahadith_sim_sorted = all_ahadith_sim.sort(key=lambda x: x[1], reverse=True)
# top_results = all_ahadith_sim_sorted[:10]
# for result in top_results:
#     print(f"ID: {result[0]}, Similarity Score: {result[1]}")

# Print 10 most similar ahadith in a format
print("Top 10 similar Ahadith:")
for score, i in all_ahadith_sim_sorted[0:10]:
    print("{} \t {:.4f} \t {}".format(documents_b_m[i], hadith_cos_sim[0][i],metadatas_b_m[i]))
```

```python
# Make an empty dataframe to store data from array
top_10_relevant_ahadith=pd.DataFrame(columns=["Hadith Source","Cos_Similarity","Hadith_en"])
print(top_10_relevant_ahadith)
```

```python
# Store data from array to empty dataframe
for score, i in all_ahadith_sim_sorted[0:10]:
    new_row = {"Hadith Source": metadatas_b_m[i], "Cos_Similarity": hadith_cos_sim[0][i], "Hadith_en": documents_b_m[i]}
    # new_row = {"Hadith Book": bukhari__muslim_df.source[i],"Hadith Number": metadatas_b_m[i], "Cos_Similarity": hadith_cos_sim[0][i], "Hadith_en": documents_b_m[i]}
    df2 = pd.DataFrame([new_row])
    top_10_relevant_ahadith = pd.concat([top_10_relevant_ahadith, df2], ignore_index=True)
top_10_relevant_ahadith
```

```python
# Save dataframe to a new csv as result file
top_10_relevant_ahadith.to_csv('top_10_relevant_ahadith5.csv', index=False)
```
